```python
## House Price Prediction Example
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r"data\house_data.csv")
```

```python
df.head()
```

```python
df.corr()
```

```python
important_features = ['price','bedrooms','bathrooms','floors','waterfront','grade','lat','long']
data = df[important_features]
```

```python
data.head()
```

```python
data.describe()
```

```python
#Chick pre-processing
data.info()
```

```python
X = data.iloc[:, 1:].values
Y = data.iloc[:, 0].values
```

```python
X
```

```python
Y
```

```python

from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python

from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

print("b = ", lin_reg.intercept_)
print("a = ", lin_reg.coef_)
```

```python
y_predicted = lin_reg.predict(x_test)
```

```python
y_predicted
```
